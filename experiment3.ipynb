```python
#whether given number is armstrong or not
n = int(input())
s = 0
digits = 0
for _ in str(n):
    digits += 1
for d in str(n):
    s += int(d) ** digits
print("Armstrong" if s == n else "Not Armstrong")
```

```python
#fibonacci series
n = int(input("Enter number of terms: "))
a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b

```

```python
#given no is prime or not
num=int(input("Enter a number:"))
if num <= 1:
    print("Not Prime")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

```python
#given number is palindrome or not
n = int(input("Enter a number: "))
temp = n
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
if temp == rev:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
#print sum of digits
num=int(input("Enter a number:"))
sum=0
while num>0:
    digit=num%10
    sum=sum+digit
    num=num//10
print("sum of numbers:",sum)
```

```python
n = 5

for i in range(n):
    # left numbers
    for j in range(1, n - i + 1):
        print(j, end="")
    # stars
    for j in range(i):
        print("*", end="")
    # stars again (middle)
    for j in range(i):
        print("*", end="")
    # right numbers
    for j in range(n - i, 0, -1):
        print(j, end="")
    print()
```
