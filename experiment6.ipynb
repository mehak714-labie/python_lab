```python
#i. Write a Python function to find the maximum and minimum numbers from a sequence of numbers
def max_min(num):
    if not num:                       # Handle empty list case
        return None, None
    max = num[0]
    min = num[0]
    for n in num:
        if n > max:
            max = n
        if n < min:
            min = n
    return max, min
numstr = input("Enter numbers (separated by spaces): ")  # Get numbers as a space-separated string
numl = [int(x) for x in numstr.split()] # Convert the string of numbers into a list of integers
max, min = max_min(numl)
if max is not None:
    print("Maximum number is", max)
    print("Minimum number is", min)
else:
    print("No numbers were entered.")
```

```python
#ii. Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number
def sum_of_cubes(num):
    sum = 0
    for i in range(1, num):
        sum += i**3
    return sum
num=int(input("Enter a positive integer: "))
if num <= 0:
  print("Please enter a positive integer.")
else:
  result = sum_of_cubes(num)
print("The sum of the cubes of all the positive integers smaller than", num, "is", result)
```

```python
#iii. Write a Python function to print 1 to n using recursion.
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)
n=int(input("Enter a positive integer: "))
if n <= 0:
  print("Please enter a positive integer.")
else:
  print("The numbers from 1 to", n, "are:")
print_numbers(n)
```

```python
#iv. Write a recursive function to print Fibonacci series upto n terms.
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib_list = fibonacci(n - 1)
        fib_list.append(fib_list[-1] + fib_list[-2])
        return fib_list
n=int(input("Enter a positive integer: "))
if n <= 0:
    print("Please enter a positive integer.")
else:
    fib_list = fibonacci(n)
    print("The Fibonacci series upto", n, "terms is:", fib_list)
```

```python
#v. Write a lambda function to find volume of cone.
import math
cone_volume = lambda r, h: (1/3) * math.pi * r**2 * h
r = float(input("Enter the radius of the cone: "))
h = float(input("Enter the height of the cone: "))
volume = cone_volume(r, h)
print("The volume of the cone is:", volume)
```
