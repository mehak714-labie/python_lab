```python
#vi. Write a lambda function which gives tuple of max and min from a list.
max_min = lambda lst: (max(lst), min(lst))      #max () and min() built in functions
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    lst.append(int(input("Enter element: ")))   #adding elements in list(lst)
result = max_min(lst)
print("Maximum and Minimum values are:", result)
```

```python
#viii. Write a program to check whether all the values in a dictionary are same or not using lambda function.

n = int(input("Enter number of k-v pairs: "))
d = {}
for i in range(n):
    key = input("Enter key: ")
    value = input("Enter value: ")
    d[key] = value
# Lambda function to check if all values are same
check = lambda x: len(set(x.values())) == 1
if check(d):
    print("All values are same.")
else:
    print("All values are not same.")
```

```python
#ix. Write a program to create two lists and generate a dictionary with keys from list1 and values from list2.
n = int(input("Enter number of elements: "))
l1 = []
l2 = []
print("Enter elements for List 1 (keys):")
for i in range(n):
    l1.append(input())
print("Enter elements for List 2 (values):")
for i in range(n):
    l2.append(input())
dictionary = dict(zip(l1, l2))
print("Generated Dictionary:")
print(dictionary)
```
