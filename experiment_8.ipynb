```python
#tkinter window.
import tkinter as tk
import threading
def run():
    win = tk.Tk()
    win.title("My First Window")
    win.geometry("400x300")
    win.resizable(False, False)
    tk.Label(win, text="Welcome!", font=("Arial", 20)).pack(pady=100)
    win.mainloop()
threading.Thread(target=run).start()
```

```python
#basic calculator.
import tkinter as tk
import threading
def run():
    def calculate(op):
        try:
            a, b = float(e1.get()), float(e2.get())
            ops = {'+': a+b, '-': a-b, '*': a*b, '/': a/b}
            res.config(text=f"Result: {ops[op]}")
        except ZeroDivisionError:
            res.config(text="Error: Division by zero")
        except ValueError:
            res.config(text="Error: Invalid input")
    win = tk.Tk()
    win.title("Calculator")
    win.geometry("300x250")
    tk.Label(win, text="Number 1:").pack()
    e1 = tk.Entry(win); e1.pack()
    tk.Label(win, text="Number 2:").pack()
    e2 = tk.Entry(win); e2.pack()
    f = tk.Frame(win); f.pack(pady=10)
    for op in ['+','-','*','/']:
        tk.Button(f, text=op, width=5,
                  command=lambda o=op: calculate(o)).pack(side=tk.LEFT)
    res = tk.Label(win, text="Result: "); res.pack()
    win.mainloop()
threading.Thread(target=run).start()
```

```python
#student registration.
import tkinter as tk
import sqlite3
# DB setup
conn = sqlite3.connect("students.db")
conn.execute("""CREATE TABLE IF NOT EXISTS students
             (name TEXT, course TEXT, email TEXT)""")
conn.commit()
def register():
    conn.execute("INSERT INTO students VALUES (?,?,?)",
                 (name_var.get(), course_var.get(), email_var.get()))
    conn.commit()
    status.config(text="Registered Successfully!")
win = tk.Tk()
win.title("Student Registration")
win.geometry("350x250")
name_var = tk.StringVar()
course_var = tk.StringVar()
email_var = tk.StringVar()
for label, var in [("Name", name_var), ("Course", course_var), ("Email", email_var)]:
    tk.Label(win, text=label).pack()
    tk.Entry(win, textvariable=var).pack()
tk.Button(win, text="Register", command=register).pack(pady=10)
status = tk.Label(win, text="")
status.pack()

win.mainloop()
```

```python
#task manager.
import tkinter as tk
import sqlite3
conn = sqlite3.connect("tasks.db")
conn.execute("CREATE TABLE IF NOT EXISTS tasks (id INTEGER PRIMARY KEY, task TEXT)")
conn.commit()
def load_tasks():
    listbox.delete(0, tk.END)
    for row in conn.execute("SELECT id, task FROM tasks"):
        listbox.insert(tk.END, f"{row[0]}: {row[1]}")
def add_task():
    conn.execute("INSERT INTO tasks (task) VALUES (?)", (task_var.get(),))
    conn.commit()
    load_tasks()
def delete_task():
    selected = listbox.get(tk.ACTIVE)
    if selected:
        task_id = selected.split(":")[0]
        conn.execute("DELETE FROM tasks WHERE id=?", (task_id,))
        conn.commit()
        load_tasks()
def edit_task():
    selected = listbox.get(tk.ACTIVE)
    if selected:
        task_id = selected.split(":")[0]
        conn.execute("UPDATE tasks SET task=? WHERE id=?",
                     (task_var.get(), task_id))
        conn.commit()
        load_tasks()
win = tk.Tk()
win.title("Task Manager")
win.geometry("400x350")
task_var = tk.StringVar()
tk.Entry(win, textvariable=task_var, width=40).pack(pady=5)
frame = tk.Frame(win)
frame.pack()
tk.Button(frame, text="Add", command=add_task).pack(side=tk.LEFT, padx=5)
tk.Button(frame, text="Edit", command=edit_task).pack(side=tk.LEFT, padx=5)
tk.Button(frame, text="Delete", command=delete_task).pack(side=tk.LEFT, padx=5)
listbox = tk.Listbox(win, width=50, height=10)
listbox.pack(pady=10)
load_tasks()
win.mainloop()
```

```python
#login and signup authentication system.
import tkinter as tk
import sqlite3
conn = sqlite3.connect("users.db")
conn.execute("CREATE TABLE IF NOT EXISTS users (username TEXT, password TEXT)")
conn.commit()
def signup():
    u, p = user_var.get(), pass_var.get()
    exists = conn.execute("SELECT * FROM users WHERE username=?", (u,)).fetchone()
    if exists:
        status.config(text="User already exists!")
    else:
        conn.execute("INSERT INTO users VALUES (?,?)", (u, p))
        conn.commit()
        status.config(text="Signup successful!")
def login():
    u, p = user_var.get(), pass_var.get()
    user = conn.execute("SELECT * FROM users WHERE username=? AND password=?",
                        (u, p)).fetchone()
    status.config(text="Login successful! Welcome " + u if user else "Invalid credentials!")
win = tk.Tk()
win.title("Login / Signup")
win.geometry("300x220")
user_var = tk.StringVar()
pass_var = tk.StringVar()
tk.Label(win, text="Username:").pack()
tk.Entry(win, textvariable=user_var).pack()
tk.Label(win, text="Password:").pack()
tk.Entry(win, textvariable=pass_var, show="*").pack()
frame = tk.Frame(win)
frame.pack(pady=10)
tk.Button(frame, text="Login", command=login).pack(side=tk.LEFT, padx=10)
tk.Button(frame, text="Signup", command=signup).pack(side=tk.LEFT, padx=10)
status = tk.Label(win, text="")
status.pack()
win.mainloop()
```
