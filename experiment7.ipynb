```python
#i. Add few names, one name in each row, in “name.txt file”.a. Count no of names b. Count all names starting with vowelb c. Find longest name
from google.colab import files
uploaded = files.upload()
with open("name.txt", "r") as f:
    names = f.readlines()
names = [name.strip() for name in names]    #removes newline issues in count
# a) Count total no of names
tnames = len(names)
print("Total number of names:", tnames)
# b) Count names starting with vowel
vowels = ('A', 'E', 'I', 'O', 'U')
vo_count = 0   #vowel counttt
for name in names:
    if name.upper().startswith(vowels):
        vo_count += 1
print("Names starting with vowel:", vo_count)
# c) Find longest name
long_name = max(names, key=len)                  #key (what basis the elements should be compared or sorted)
print("Longest name:", long_name)
```

```python
#ii. Store integers in a file.a. Find the max numberb. Find average of all numbers c. Count number of numbers greater than 100
from google.colab import files
uploaded = files.upload()
with open("numbers.txt", "r") as file:
    num = [int(line.strip()) for line in file]
# a) Find maximum number
max_num = max(num)
print("Maximum number:", max_num)
# b) Find average
average = sum(num) / len(num)
print("Average:", average)
# c) Count numbers greater than 100
count = 0
for num in num:
    if num > 100:
        count += 1
print("Numbers greater than 100:", count)
```

```python
#iii. Assume a file city.txt with details of 5 cities in given format (cityname population(in lakhs) area(in sq KM) a. Display details of all cities b. Display city names with population more than 10Lakhs c. Display sum of areas of all cities
from google.colab import files
uploaded = files.upload()
with open("city.txt", "r") as file:
    cities = file.readlines()
total_area = 0
print("Details of all cities:\n")
for line in cities:
    data = line.split()        # Split line into parts
    cname = data[0]              #ciyname     #0-name 1-pop 2-area
    population = float(data[1])
    area = float(data[2])
    print("City:", cname)                   # a) Display details
    print("Population (Lakhs):", population)
    print("Area (sq km):", area)
                                            # b) Population more than 10 lakhs
    if population > 10:
        print(cname, "has population more than 10 Lakhs")
                                             # c) Add area for total
    total_area += area
print("\nTotal area of all cities:", total_area)
```

```python
#iv. Input two values from user where the first line contains N, the number of test cases.The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.
# Read number of test cases
N = int(input("Enter number of test cases: "))

for i in range(N):
    try:
        a, b = input("Enter two numbers separated by space: ").split()
        a = int(a)
        b = int(b)
        result = a // b
        print("Result:", result)
    except ZeroDivisionError as e:
        print("Error:", e)
    except ValueError as e:
        print("Error:", e)
```

```python
#v. Create multiple suitable exceptions for a file handling
from google.colab import files
uploaded = files.upload()
try:
    fname = list(uploaded.keys())[0]   # Automatically get uploaded file name
    with open(fname, "r") as file:
        content = file.read()
        print("File Content:", content)
        number = int(content)
        print("Number in file:", number)
except FileNotFoundError:
    print("Error: File not found.")
except PermissionError:
    print("Error: Permission denied.")
except ValueError:                                           #not an integer
    print("Error: File contains invalid data .")
except Exception as e:
    print("Unexpected Error:", e)
```

```python
#vi. Write a program to create a counter to show how many times the program is executed.
filename = "/content/counter.txt"
try:
    with open(filename, "r") as file:
        count = int(file.read())
except FileNotFoundError:
    count = 0
count += 1
with open(filename, "w") as file:
    file.write(str(count))

print("Program has been executed", count, "times.")
```
