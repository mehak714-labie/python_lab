```python
x=int(input("enter number1:"))
y=int(input("enter number2:"))
ch=int(input("enter ur choice:"))
if ch==1:
    z = x+y
    print("addition:", z)
elif ch==2:
    z = x-y
    print("subtraction:", z)
elif ch==3:
    z = x*y
    print("multiplication:", z)
else:
    z = x/y
    print("division:", z)
```

```python
import math
x=int(input("enter side1:"))  #sides are x,y,z
y=int(input("enter side2:"))
z=int(input("enter side3:"))
s=(x+y+z)/2
a=math.sqrt(s*(s-x)*(s-y)*(s-z))  #herons formula
print("the area of triangle is",a)
```

```python
secs = int(input("Enter total seconds:"))
hrs = secs // 3600
rem_sec = secs % 3600
minutes = rem_sec // 60
remaining_seconds = rem_sec % 60
print("Hours:", hrs)
print("Minutes:", minutes)
print("Seconds:", remaining_seconds)
```

```python
print("A B | A&B A|B A^B")
for a in [0, 1]:
    for b in [0, 1]:
        print(a, b, "|", a & b, " ", a | b, " ", a ^ b)
```
