```python
a = float(input("Enter a: "))     #Q5 Check whether the quadratic equation has real roots or imaginary roots. Display the roots.
b = float(input("Enter b: "))
c = float(input("Enter c: "))
d = b*b - 4*a*c   # discriminant
if d > 0:
    r1 = (-b + d**0.5) / (2*a)
    r2 = (-b - d**0.5) / (2*a)
    print("Roots are real and distinct")
    print("Root 1:", r1)
    print("Root 2:", r2)
elif d == 0:
    r = -b / (2*a)
    print("Roots are real and equal")
    print("Root:", r)
else:
    real = -b / (2*a)
    imag = (abs(d)**0.5) / (2*a)
    print("Roots are imaginary")
    print("Root 1:", real, "+", imag, "i")
    print("Root 2:", real, "-", imag, "i")
```

```python
print("STUDENT GRADE SHEET")
subjects = ["PDS", "Physics", "Chemistry", "English", "Python"]
total = 0
for subject in subjects:
    marks = int(input(f"Enter marks for {subject} (out of 100): "))
    total += marks
percentage = (total / 500) * 100
cgpa = percentage / 10
print("\nRESULT")
print(f"Total Marks : {total} / 500")
print(f"Percentage  : {percentage:.2f}%")
print(f"CGPA        : {cgpa:.2f}")
if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:

    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "Invalid"
print(f"Grade       : {grade}")
```
